bubble sort

- worst and average case
Time complexity - O(n²)
Space complexity - O(1)



```python
nums = [5,1,6,8,2,4,9]

def bubble_sort(arr):
    n = len(arr)
    
    for i in range(n-2,-1,-1):
        for j in range(0,i+1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                
    return arr

print(bubble_sort(nums))
```

```python
# best case
# Time complexity - O(n)
# Space complexity - O(1)

nums = [5,1,6,8,2,4,9]

def bubble_sort(arr):
    n = len(arr)
    
    for i in range(n-2,-1,-1):
        is_swap = False
        for j in range(0,i+1):
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                is_swap = True
        if is_swap == False:
            break
                
    return arr

print(bubble_sort(nums))
```
