n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

constraints:
1<=n[i]<=10

n can have 10⁸ elements

m can have 10⁸ elements

```python
# Bruteforce method

n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

count = dict()

for num in m:
    count[num]=0
    for x in n:
        if x==num:
            count[num]+=1
            
print(count)

# in worst case n will have 10⁸ elements and m also
# so time complexity will be 10¹⁶ - TLE error will occur
```

```python
# optimal solution

n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

hash_list = [0] * 11
for num in n:               #tc O(n)
    hash_list[num]+=1
    
for num in m:               #tc O(m)
    if num<1 or num>10:
        print(0)
    else:
        print(hash_list[num])
        
# Time complexity - O(n+m)
# Space complexity - O(1)   - becuase we used hash_list to store but its size will remain constant
```

```python
# problem solution using dictionary

n = [5,3,2,2,1,5,5,7,5,10]
m = [10,111,1,9,5,67,2]

hash_list = {x:0 for x in range(1,11)}

for x in n:
    if x in hash_list:
        hash_list[x]+=1
        
# for num in m:
#     if num in hash_list:
#         print(hash_list[num])
#     else:
#         print(0)

res = {x:hash_list[x] for x in m if x in hash_list}
print(res)
```

```python
# char hashing

s = 'azyxyyzaaaa'
q = ['d','a','y','x']

hash_list = {}

for ch in s:
    ascii_value = ord(ch)
    idx = ascii_value - 97
    hash_list[idx] = hash_list.get(idx,0)+1
    
res = {ch:hash_list.get(ord(ch)-97,0) for ch in q}
print(res)
```
