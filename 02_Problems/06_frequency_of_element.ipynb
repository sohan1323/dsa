method 1

Time complexity O(n)

```python
nums = [5,6,7,7,1,9,111,1,1,5,1,1]

def counter(lst):
    freq = {}
    for i in lst:                   #O(n)
        if i not in freq.keys():    #O(1)
            freq[i]=1
        else:
            freq[i]+=1              #O(1)
            
    return freq

print(counter(nums))
```

method 2

Time complexity - O(1)

```python
nums = [5,6,7,7,1,9,111,1,1,5,1,1]

def counter(lst):
    hash_map = dict()
    n = len(lst)
    
    for i in range(0,n):
        hash_map[lst[i]] = hash_map.get(lst[i],0)+1  # dict.get(n,0) -> if key exist return value of key n, else return 0
        
    return hash_map

print(counter(nums))
```
