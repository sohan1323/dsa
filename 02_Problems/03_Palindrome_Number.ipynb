```python
#time complexity - O(log₁₀(N)) -> in following solution we are dividing by 10 so it is base 10, if we divide by suppose 7 it will base 7

n = 1232

num = n
rev_num = 0

while num > 0:
    rem = num % 10
    rev_num = rev_num*10 + rem
    num = num // 10

print(rev_num)

if n == rev_num:
    print("yes it is palindrome")
else:
    print('its not palindrome')

```
