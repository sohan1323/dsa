# String palindrome check using recursion

This notebook checks whether a string is a palindrome by recursively comparing characters from the ends of the string toward the center.

Time complexity: O(n). Space complexity: O(n) recursion stack.

```python
s = "noon"

def p(s,left,right):
    if left>=right:
        return True
    if s[left]!=s[right]:
        return False
    
    return p(s,left+1,right-1)

if p(s,0,len(s)-1):
    print("y")
else:
    print("n")
```
