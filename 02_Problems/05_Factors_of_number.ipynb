BruteForce
Total numbers = n
Time complexity = O(n)  [modulo and append are O(1)]
Space complexity = O(k) -> k would be total no. of factors

```python
def factor_finder(num):
    
    fact = []
    
    for i in range(1,num+1):
        if num%i==0:
            fact.append(i)
            
    return fact
        
factor_finder(10)
```

Better solution - on carefull observation factors are only in first half (n//2) in second half only number itself is the factor, so we need to loop only halfway and append number itself at last

Time complexity - O(n/2) ~ O(n)
Space complexity - O(k)

```python
def factor_finder(num):
    fact = []
    for i in range(1,num//2+1):
        if num%i==0:
            fact.append(i)
    fact.append(num)
    
    return fact

print(factor_finder(10))
```

Optimal solution
suppose num is 36
then we find factors
1 is factor and 36//1 = 36  ---> (1,36)
2 is factor and 36//2 = 18  ---> (2,18)
3 is factor and 36//3 = 12  ---> (3,12)
like this ...
so we need to find only upto square root of the number

Time complexity - O(√n)  ---> if sorted => time complexity of sorting is O(n log n)
Space complexity - O(k)

```python
def factor_finder(num):
    fact = []
    for i in range(1,(int(num**0.5))+1):
        if num%i==0:
            fact.append(i)
            if num//i!=i:
                fact.append(num//i)
    # fact.sort()
    return fact

print(factor_finder(36))
        
        
```
