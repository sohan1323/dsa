# Finding factors of a number

This notebook shows multiple factor-finding methods: brute-force iteration up to n, improved iteration up to n/2, and the optimal divisor-pair method using the square root of n. It explains when each approach is appropriate.

Complexities:
- brute force: O(n)
- halfway check: O(n/2) ~ O(n)
- optimal divisor pairing: O(√n)
Space complexity: O(k) for k factors.

```python
def factor_finder(num):
    
    fact = []
    
    for i in range(1,num+1):
        if num%i==0:
            fact.append(i)
            
    return fact
        
factor_finder(10)
```

```python
def factor_finder(num):
    fact = []
    for i in range(1,num//2+1):
        if num%i==0:
            fact.append(i)
    fact.append(num)
    
    return fact

print(factor_finder(10))
```

```python
def factor_finder(num):
    fact = []
    for i in range(1,(int(num**0.5))+1):
        if num%i==0:
            fact.append(i)
            if num//i!=i:
                fact.append(num//i)
    # fact.sort()
    return fact

print(factor_finder(36))
        
        
```
