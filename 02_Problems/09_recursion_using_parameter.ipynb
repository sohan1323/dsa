# Recursion with parameters

This notebook shows recursive functions that take parameters to control repetition and ordering. It includes printing a value x n times and recursively printing numbers from 1 to n.

Time complexity: O(n). Space complexity: O(n) due to recursion stack.

```python
# print x, n times
```

```python
def func(x,n):
    if n==0:
        return
    print(x)
    func(x,n-1)
    
func(15,4)
```
