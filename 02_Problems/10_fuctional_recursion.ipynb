# Functional recursion example

This notebook demonstrates a recursive function that returns values instead of only printing them. It computes the sum of integers from 1 to n by returning partial results through recursive calls.

Time complexity: O(n). Space complexity: O(n) recursion stack.

```python
def func(n):
    if n == 1:
        return 1
    return n+func(n-1)

print(func(5))
```
