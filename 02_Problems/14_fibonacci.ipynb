# Fibonacci sequence with recursion

This notebook builds Fibonacci numbers using recursion. It includes a commented naive recursive version and a more efficient recursive list-building version that constructs the sequence iteratively through recursion.

Time complexity: O(n^2) for the recursive list-building version because each call copies a list. Space complexity: O(n^2) due to the accumulated list and call stack. The naive version would be O(2^n) time and O(n) space.

```python
# normal
# def func(num):
#     if num==0 or num==1:
#         return num
#     return func(num-1)+func(num-2)

# answer = [func(i) for i in range(8)]
# print(answer)

# efficient
def fib_list(n):
    if n == 0:
        return [0]
    if n == 1:
        return [0, 1]
    
    res = fib_list(n-1)
    res.append(res[-1] + res[-2])
    return res

print(fib_list(7))
```
