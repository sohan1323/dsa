print "hello" 4 times

```python
count = 0

def func():             # head recurssion - first done job then call function
    global count
    if count == 4:
        return
    print("hello")      # this job
    count+=1
    func()              # this call
    
func()
    
```

```python
count = 0

def func():         # tail recurssion - first call function then do job
    global count
    if count == 4:
        return
    count+=1
    func()          # this call
    print("hello")  # this job
    
func()
```

time complexity - O(n+1) ~ O(n)
space complexity - O(n+1) ~ O(n)
