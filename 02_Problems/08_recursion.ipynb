# Recursion examples: head vs tail recursion

This notebook demonstrates recursion patterns by printing a message multiple times. It shows head recursion (work after the recursive call) and tail recursion (work before the recursive call), and highlights the recursion stack cost.

Time complexity: O(n). Space complexity: O(n) because of recursion stack depth.

```python
# python default recurssion limit - 987 calls
```

```python
count = 0

def func():         # head recurssion - first call function then do job
    global count
    if count == 4:
        return
    count+=1
    func()          # this call
    print("hello")  # this job
    
func()
```

```python
count = 0

def func():             # tail recurssion - first done job then call function
    global count
    if count == 4:
        return
    print("hello")      # this job
    count+=1
    func()              # this call
    
func()
    
```
