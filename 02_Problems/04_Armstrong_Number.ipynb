# Armstrong number check

This notebook determines whether a number is an Armstrong number by summing each digit raised to the power of the digit count and comparing the result to the original number.

Time complexity: O(d) for d digits. Space complexity: O(1).

```python
# 153 = 1³ + 5³ + 3³  -> number is three digit so power is raised to 3
#     = 1 + 125 + 27
#     = 153
```

```python
n = 1634

cnt = num = n
sum = 0
count = 0
# power = len(str(num))

while cnt>0:
    count+=1
    cnt = cnt//10

while num>0:
    rem = num%10
    sum += (rem**count)
    num = num//10
    
if n == sum:
    print("its armstrong number")
else:
    print("not a armstrong number")
```
