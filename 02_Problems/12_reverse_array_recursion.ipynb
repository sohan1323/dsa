# Reverse an array with recursion

This notebook reverses a list in-place using recursion by swapping elements from the ends toward the center. The recursion reduces the active subarray on each step.

Time complexity: O(n). Space complexity: O(n) recursion stack.

```python
num = [5,7,3,2,6,1,5,9]

def func(arr,left,right):
    if left>=right:
        return
    arr[left],arr[right] = arr[right],arr[left]
    func(arr,left+1,right-1)
    
func(num,0,len(num)-1)
print(num)
```
