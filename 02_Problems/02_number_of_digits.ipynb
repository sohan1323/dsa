```python
#time complexity - O(log₁₀(N)) -> in following solution we are dividing by 10 so it is base 10, if we divide by suppose 7 it will base 7
n = 6443522
num = n
count = 0

while num>0:
    count+=1
    num = num//10
    
print(count)

#method 2 - log method
import math

def countDigit(num1):
    return int(math.log10(num1))+1

print(countDigit(n))
```
