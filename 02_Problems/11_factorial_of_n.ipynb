# Factorial using recursion

This notebook computes the factorial of a number recursively. It handles base cases for 0 and 1, and recursively multiplies n by factorial(n-1).

Time complexity: O(n). Space complexity: O(n) recursion stack.

```python
def factorial(n):
    if n<0:
        return f"enter positive int"
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)

factorial(-3)
```
